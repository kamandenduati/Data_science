```python
import pandas as pd
import numpy as np
import seaborn as sns
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Survey = sorted(glob('file/IT*.csv'))
```

```python
survey=pd.concat((pd.read_csv(file) for file in Survey),ignore_index=True)
```

```python
survey.head()
```

```python
survey.info()
```

```python
survey.isnull().sum()
```

```python
survey.dropna()
```

```python
survey.describe()
```

```python
sns.histplot(data=survey,x='Current Salary')
plt.title("Current Salary")
plt.show()
```

```python
sns.histplot(data=survey,x='Salary one Timestamp ago')
plt.title("Salary one Timestamp ago")
plt.show()
```

```python
sns.histplot(data=survey,x='Salary two Timestamps ago')
plt.title("Salary two Timestamps ago")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=survey, x='Current Salary', y='Gender')
plt.title('Distribution of Salary by Gender')
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(data=survey, x='Age', y='Current Salary')
plt.title('Distribution of Salary by age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=survey, x='Years of experience', y='Current Salary')
plt.title('Salary vs. Years of Experience')
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = survey.corr()
sns.heatmap(corr, cmap='viridis', annot=True)
plt.title('Correlation matrix')
plt.show()
```
